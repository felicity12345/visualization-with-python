```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\Felicity\Desktop\Tutoria\Ice Cream Ratings.csv")
df = df.set_index('Date')
df 
```

```python
print(plt.style.available)
plt.style.use('fivethirtyeight')
```

```python
df.plot(kind = 'line', title = 'Ice Cream Ratings', xlabel = 'Daily Ratings', ylabel = 'Scores')
```

```python
df.plot.barh(stacked = True, title = 'Ice Cream Ratings')
```

```python
df.plot.scatter(x = 'Flavor Rating', y = 'Overall Rating', s = 300, c = 'green', title = 'Ice Cream Ratings')
```

```python
df.plot.hist(bins = 10, title = 'Ice Cream Ratings')
```

```python
df.boxplot()
```

```python
df.plot.area(figsize = (6, 3), title = 'Ice Cream Ratings')
```

```python
df.plot.pie(y = 'Flavor Rating', figsize = (7, 7), title = 'Ice Cream Ratings')
colors = ['skyblue', 'lightgreen', 'lightcoral', 'lightsalmon'] 
explode = (0.1, 0, 0, 0)
```
